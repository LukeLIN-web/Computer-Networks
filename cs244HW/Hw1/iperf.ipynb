```python
import subprocess
import re
import matplotlib.pyplot as plt
window_sizes = ["16K","32K","64K", "128K", "256K", "512K", "1024K"]

server_ip = "mcnode36"
```

# Bandwidth vs. window size

```python
bandwidth_values = []
for window_size in window_sizes:
    cmd = [f"iperf -c {server_ip} -w {window_size}"]

    result = subprocess.check_output(cmd, shell=True,text=True)

    bandwidth_match = re.findall(r"(?:\d*\.*\d+)",  result)

    if bandwidth_match:
        bandwidth = float(bandwidth_match[-1])
        bandwidth_values.append(bandwidth)
        print(f"Window Size: {window_size}, Bandwidth: {bandwidth} Mbits/sec")

plt.figure(figsize=(8, 6))
plt.plot(window_sizes, bandwidth_values, marker='o', linestyle='-')
plt.title('Bandwidth vs. Window Sizes')
plt.xlabel('Window Sizes')
plt.ylabel('Bandwidth (Mbits/sec)')
plt.grid(True)
plt.show()
```

# Bandwidth vs. parallel connection
setting:  window size = 128KBytes(default)

```python
bandwidth_values = []
parallel_connections = [1, 2, 4, 8,16,32,64]

for connections in parallel_connections:
    cmd = [f"iperf -c {server_ip} -P {connections}"]

    result = subprocess.check_output(cmd, shell=True,text=True)

    bandwidth_match = re.findall(r"(?:\d*\.*\d+)",  result)

    if bandwidth_match:
        bandwidth = float(bandwidth_match[-1])
        bandwidth_values.append(bandwidth)
        print(f"Parallel Connections: {connections}, Bandwidth: {bandwidth} Mbits/sec")

plt.figure(figsize=(14, 7))
plt.plot(parallel_connections, bandwidth_values, marker='o', linestyle='-')
plt.title('Bandwidth vs. Parallel number')
plt.xlabel('Parallel number')
plt.ylabel('Bandwidth (Mbits/sec)')
plt.grid(True)
plt.show()
```
